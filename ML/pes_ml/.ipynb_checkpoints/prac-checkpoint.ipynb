```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 

xAxis = np.array([1,2,3,4,5])
yAxis = np.array([3,4,5,6,7])


#basic adding label header
plt.title("demo data")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(xAxis, yAxis, 'D:r')
plt.show()


```

```python
#grid types
plt.plot(xAxis, yAxis, 'o:r')
plt.grid()
plt.grid(axis='y')
```

```python
plt.subplot(1,2, 1)
plt.plot(xAxis, yAxis)

plt.subplot(1,2,2)
plt.plot(xAxis, yAxis)

plt.show()
```

graphs or charts

```python
#scatter and to give the color array and the syntax
#syntax for color we can give one color for all or else it should match the number of columns 
#ie n columns and n colors
plt.scatter(xAxis, yAxis, c=['red', 'green', 'blue', 'purple', 'brown'])
plt.show()
```

```python
#color map is a map that ranges from range specified
#color map is also called viridis
#syntax 
# plt.scatter(x, y, c=colors, cmap='viridis')

color = np.array([0, 10, 20, 30, 40])

plt.scatter(xAxis, yAxis, c=color, cmap='viridis')

#this command is displaying the color bar aside
plt.colorbar()

plt.show()
```

```python
#bar chart
#here the bar plot can take categorical data in x axis and numerical data in y axis and both categorical data is also allowed
#but note it allows only same elements as x axis and y axis i.e both size should be same
xAxis = np.array(["pradeep", "vinuyak", "bisnu", "dhanush"])
#plt.bar is vertical visualisation of data and plt.barh is horizontal visualisation of data
yAxis = np.array([1, 2, 3, 4])
#make use of height attribute if plt.barh and make use of width attribute if bar 
plt.barh(xAxis, yAxis, height=0.3)
plt.show()
```

```python
#histogram
#make use of hist function to create histogram
#this show the data that are dispersed at a point singe arraay as a parameter or a column in a dataframe

#normal function parameter
# last parameter is the size of the list 
#loc: This represents the mean or average value of the distribution. In this case, it is set to 170, indicating that the distribution is centered around 170.

# scale: This parameter represents the standard deviation of the distribution.
# It determines the spread or dispersion of the generated values. In the code snippet, the scale is set to 10,
# indicating that the values will vary around the mean with a standard deviation of 10.

# size: This specifies the shape or size of the output array. In this case, it is set to 250, indicating that the generated array will have 250 elements. 
x = np.random.normal(170, 10, 10)
plt.hist(x)
plt.show()
```

```python
#same with the pie chart colors are random 
#moves counterclock wise
#labels can also be provided
label = np.array(["rad", "blue", "cyan", "green", "date", "rad", "blue", "cyan", "green", "date"])
plt.pie(x, labels=label)
plt.show()
```
