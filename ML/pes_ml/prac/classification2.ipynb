```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('/home/pilli007/Documents/pes_sem2/ML/pes_ml/datasets/train.csv')
test_data = pd.read_csv('/home/pilli007/Documents/pes_sem2/ML/pes_ml/datasets/test.csv')

train_data
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

encoded = train_data.iloc[:, :-2]

for i in encoded:
    train_data[i] = le.fit_transform(train_data[[i]])
    test_data[i] = le.fit_transform(test_data[[i]])
```

```python
xTrain = pd.DataFrame(train_data.drop(columns=['class', 'Id']))
yTrain = pd.DataFrame(train_data['class'])

xTest = test_data.iloc[:, 0:-1]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rfc = RandomForestClassifier(n_estimators=50, random_state=40)

rfc.fit(xTrain, yTrain)
```

```python
yPred = rfc.predict(xTest)

op = pd.DataFrame({'id': test_data['Id'], 'class': yPred})

data = op.to_csv('pes1pg22ca145.csv', index=False)

op.head(10)
```
