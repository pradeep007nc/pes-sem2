```python
import pandas as pd

df = pd.read_csv('/home/pilli007/Documents/pes_sem2/ML/pes_ml/datasets/tic_tac_toe_dataset.csv')

test = pd.read_csv('/home/pilli007/Documents/pes_sem2/ML/pes_ml/datasets/test.csv')

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df.columns

arr = ['top_left_square', 'top_middle_square', 'top_right_square',
       'middle_left_square', 'middle_middle_square', 'middle_right_square',
       'bottom_left_square', 'bottom_middle_square', 'bottom_right_square']

for i in arr:
    df[i] = le.fit_transform(df[[i]])
```

```python

X = df.iloc[:, 0:-1]
y = df['class']

df
```

```python
from sklearn.model_selection import train_test_split

xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size=0.2, random_state=1, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

rfc = RandomForestClassifier(n_estimators=50)

rfc.fit(xTrain, yTrain)

rfc.score(xTrain, yTrain)

rfc.score(xTest, yTest)
```

```python
yPred = rfc.predict(xTest)

print(accuracy_score(yTest, yPred))

yPred
```
