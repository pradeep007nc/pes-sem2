```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Salaries.csv')
df
```

```python
sb.countplot(data= df, x = 'rank', hue='sex')
```

```python
sb.jointplot(data= df, x='yrs.service', y= 'salary', hue='sex', kind='hist')
```

```python
#scatter plot with regression line is called regression plot
sb.regplot(data= df, x= 'yrs.service', y= 'salary')
```

```python
sb.barplot(data = df, x = 'rank', y='salary', hue='sex')
```

```python
sb.catplot(data = df, x= 'rank', y= 'salary', hue='sex', kind='violin')
#other attributes in kind attirbute is swarm box strip  and violin where the mid line is median
```
