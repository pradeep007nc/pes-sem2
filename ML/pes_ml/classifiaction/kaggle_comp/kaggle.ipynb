```python
import pandas as pd
dtrain=pd.read_csv(r"train.csv")
```

```python
dtrain.head(10)
```

```python
dtrain.isnull().sum()
```

```python
dtrain["TEMP"]=dtrain["TEMP"].fillna(method="bfill")
```

```python
lis=["DO","CONDUCTIVITY","BOD","NITRATE","FECAL"]
for i in lis:
    dtrain[i]=dtrain[i].fillna(method="bfill")
    
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dtrain["LOCATIONS"]=le.fit_transform(dtrain[["LOCATIONS"]])
dtrain["STATE"]=le.fit_transform(dtrain[["STATE"]])
```

```python
dtrainDF
```

```python
dtrain['TEMP_Suitability'] =  dtrain['TEMP'].apply(lambda x: 100 if 0 <= x <= 35 else 0)
dtrain['DO_Suitability'] =    dtrain['DO'].apply(lambda x: 100 if x > 5 else 0)
dtrain['PH_Suitability'] =    dtrain['PH'].apply(lambda x: 100 if 6.5 <= x <= 8.5 else 0)
dtrain['CONDUCTIVITY_Suitability'] = dtrain['CONDUCTIVITY'].apply(lambda x: 100 if x < 1000 else 0)
dtrain['BOD_Suitability'] = dtrain['BOD'].apply(lambda x: 100 if x < 5 else 0)
dtrain['NITRATE_Suitability'] = dtrain['NITRATE'].apply(lambda x: 100 if x < 10 else 0)
dtrain['FECAL_Suitability'] = dtrain['FECAL'].apply(lambda x: 100 if x == 0 else 0)
```

```python
dtrain['Suitability_Percentage'] = dtrain[['TEMP_Suitability', 'DO_Suitability', 'PH_Suitability', 'CONDUCTIVITY_Suitability', 'BOD_Suitability', 'NITRATE_Suitability', 'FECAL_Suitability']].mean(axis=1)
```

```python
dtrain
```

```python
lis=["TEMP_Suitability","DO_Suitability","PH_Suitability","CONDUCTIVITY_Suitability","BOD_Suitability","NITRATE_Suitability","FECAL_Suitability"]
for i in lis:
    del dtrain[i]
```

```python
dtrain
```

```python
xtrain=dtrain.iloc[:,1:10]
```

```python
xtrain
```

```python
ytrain=dtrain.Suitability_Percentage
```

```python
ytrain
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
rr= RandomForestRegressor(n_estimators=9, random_state=42)
```

```python
rr.fit(xtrain, ytrain)
```

```python
dtest=pd.read_csv(r"test.csv")
```

```python
dtest.head(10)
```

```python
dtest.isnull().sum()
```

```python
dtest["TEMP"]=dtest["TEMP"].fillna(method="bfill")
```

```python
dtest["LOCATIONS"]=le.fit_transform(dtest[["LOCATIONS"]])
dtest["STATE"]=le.fit_transform(dtest[["STATE"]])
```

```python
dtest['TEMP_Suitability'] =  dtest['TEMP'].apply(lambda x: 100 if 0 <= x <= 35 else 0)
dtest['DO_Suitability'] =    dtest['DO'].apply(lambda x: 100 if x > 5 else 0)
dtest['PH_Suitability'] =    dtest['PH'].apply(lambda x: 100 if 6.5 <= x <= 8.5 else 0)
dtest['CONDUCTIVITY_Suitability'] = dtest['CONDUCTIVITY'].apply(lambda x: 100 if x < 1000 else 0)
dtest['BOD_Suitability'] = dtest['BOD'].apply(lambda x: 100 if x < 5 else 0)
dtest['NITRATE_Suitability'] = dtest['NITRATE'].apply(lambda x: 100 if x < 10 else 0)
dtest['FECAL_Suitability'] = dtest['FECAL'].apply(lambda x: 100 if x == 0 else 0)
```

```python
dtest['Suitability_Percentage'] = dtest[['TEMP_Suitability', 'DO_Suitability', 'PH_Suitability', 'CONDUCTIVITY_Suitability', 'BOD_Suitability', 'NITRATE_Suitability', 'FECAL_Suitability']].mean(axis=1)
```

```python
lis=["TEMP_Suitability","DO_Suitability","PH_Suitability","CONDUCTIVITY_Suitability","BOD_Suitability","NITRATE_Suitability","FECAL_Suitability"]
for i in lis:
    del dtest[i]
```

```python
xtest=dtest.iloc[:,1:10]
```

```python
ytest=dtest.Suitability_Percentage
```

```python
y_pred = rr.predict(xtest)
```

```python
mae = mean_absolute_error(ytest, y_pred)
rmse = mean_squared_error(ytest, y_pred, squared=False)
r2 = r2_score(ytest, y_pred)
```

```python
print(f'Mean Absolute Error: {mae}')
print(f'Root Mean Squared Error: {rmse}')
print(f'R-squared: {r2}')
```
